```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
from mpmath import *
from numpy.linalg import inv
from sklearn import preprocessing
from scipy.stats import multivariate_normal


# code for reading the file into a dataframe and calculating the priors 
df=pd.read_csv("Dataset_2_Team_14.csv")  
#df2=pd.read_csv("Dataset_2_Team_14.csv") 
#frames = [df, df2]
#df = pd.concat(frames,ignore_index=True)


trainidx=np.random.randint(0,4500,(int)(0.70*len(df['Class_label'])))
valididx=np.random.randint(0,4500,(int)(0.15*len(df['Class_label'])))
testidx=np.random.randint(0,4500,(int)(0.15*len(df['Class_label'])))


train=pd.DataFrame(df,trainidx)
validation=pd.DataFrame(df,valididx)
test=pd.DataFrame(df,testidx)

L=np.array([[0,1,2],[1,0,1],[2,1,0]])

b=train['Class_label'].value_counts()
b=dict(b)

prior0=b[0]/len(train['Class_label'])
prior1=b[1]/len(train['Class_label'])
prior2=b[2]/len(train['Class_label'])
print(prior0,prior1,prior2)

data=train.as_matrix(columns=train.columns[0:3])#2d array in numpy this matrix contains the train data 

data0=[]
data1=[]
data2=[]

for ele in data:
    if(ele[2]==0):
         data0.append(ele[0:2])
    elif(ele[2]==1):
         data1.append(ele[0:2])
    elif(ele[2]==2):
         data2.append(ele[0:2])

            
ds0=np.array(data0)
ds1=np.array(data1)
ds2=np.array(data2)


mew0=np.sum(ds0,axis=0)
mew1=np.sum(ds1,axis=0)
mew2=np.sum(ds2,axis=0)

# calculating mews for the three classes 

mew0=mew0/np.size(ds0,0)
mew1=mew1/np.size(ds1,0)
mew2=mew2/np.size(ds2,0)
# print("these r the mews ")
# print(mew0,mew1,mew2)

#calculating the covariances 
cov0=np.matmul((ds0-mew0).T,ds0-mew0)/np.size(ds0,0)
cov1=np.matmul((ds1-mew1).T,ds1-mew1)/np.size(ds1,0)
cov2=np.matmul((ds2-mew2).T,ds2-mew2)/np.size(ds2,0)


for ele in data:
    
    eta0=prior0*np.power(np.linalg.det(cov0),-0.5)*np.exp(-0.5*(np.matmul(np.matmul(ele[0:2]-mew0,inv(cov0)),(ele[0:2]-mew0).T)))
    eta1=prior1*np.power(np.linalg.det(cov1),-0.5)*np.exp(-0.5*(np.matmul(np.matmul(ele[0:2]-mew1,inv(cov1)),(ele[0:2]-mew1).T)))
    eta2=prior2*np.power(np.linalg.det(cov2),-0.5)*np.exp(-0.5*(np.matmul(np.matmul(ele[0:2]-mew2,inv(cov2)),(ele[0:2]-mew2).T)))
    finalclass.append(findclass(np.array([eta0,eta1,eta2])))

print(finalclass)


###########################################################################################################

fig = plt.figure(figsize=(15,15))
ax = fig.gca(projection='3d')
# xmin=min(ds0[:,0].min(),ds1[:,0].min(),ds2[:,0].min())
# ymin=min(ds0[:,1].min(),ds1[:,1].min(),ds2[:,1].min())
# xmax=max(ds0[:,0].max(),ds1[:,0].max(),ds2[:,0].max())
# ymax=max(ds0[:,1].max(),ds1[:,1].max(),ds2[:,1].max())

X0 = np.arange(data[:,0].min(),data[:,0].max(),0.25)
Y0 = np.arange(data[:,1].min(),data[:,1].max(),0.25)
xx0,yy0 = np.meshgrid(X0, Y0)

pos = np.dstack((xx0,yy0))
rv0 = multivariate_normal(mew0, cov0)
Z0=rv0.pdf(pos)


pos = np.dstack((xx0,yy0))
rv1 = multivariate_normal(mew1,cov1)
Z1=rv1.pdf(pos)
# ax.plot_surface(xx1,yy1, Z1,color='b')
# ax.contourf(xx1,yy1,Z1,zdir='z', offset=-0.005,color='b')



pos = np.dstack((xx0,yy0))
rv2 = multivariate_normal(mew2, cov2)
Z2=rv2.pdf(pos)

surface0=ax.plot_surface(xx0,yy0,Z0,color='r')
z=ax.contourf(xx0,yy0,Z0,100,zdir='z',alpha=0.3,offset=-0.0005,cmap=cm.winter)
surface1=ax.plot_surface(xx0,yy0,Z1,color='b')
z=ax.contourf(xx0,yy0,Z1,100,zdir='z',alpha=0.3,offset=-0.0005,cmap=cm.Purples_r)
surface2=ax.plot_surface(xx0,yy0,Z2,color='g')
z=ax.contourf(xx0,yy0,Z2,100,zdir='z',alpha=0.3,offset=-0.0005,cmap=cm.Oranges_r)
plt.title('Red=class0,Blue=class1,Green=class2')


ax.set_zlim(-0.0005,0.0005)
ax.set_xlim(xx0.min(), xx0.max())
ax.set_ylim(yy0.min(), yy0.max())
# ax.set_zlim(-2, 2)

# ax = fig.gca(projection='3d')
ax.view_init(azim=220)
plt.show()
```
